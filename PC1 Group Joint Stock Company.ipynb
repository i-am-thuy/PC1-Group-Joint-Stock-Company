##PC1 – CÔNG TY CỔ PHẦN TẬP ĐOÀN PC1

Tổng quan về PC1

* Được thành lập năm 1963 với tên gọi ban đầu là Công ty Xây lắp đường dây và 
trạm. Năm 2016, PC1 thành công niêm yết trên sàn HOSE, đến nay vốn hóa thị trường đã đạt 8.084 tỷ. 
* Ba mảng kinh doanh mang lại lợi nhuận chính cho PC1 gồm:

     (i)	Mảng xây lắp điện.

    (ii)	Mảng bất động sản.

    (iii)	Mảng sản xuất điện.

(*) Mảng kinh doanh mới từ dự án Niken
Năm 2021, PC1 đã mua lại 52.27% cổ phần của Công ty cổ phần khoáng sản Tấn Phát – chủ sở hữu mỏ Niken - Đồng xã Quang Trung, xã Hà Trì tỉnh Cao Bằng. Theo thống kê, tổng trữ lượng mỏ này khoảng hơn 52000 tấn kim loại Niken. Việc giá Niken thế giới tăng phi mã kỳ vọng sẽ ghi nhận doanh thu đột biến cho PC1. 


* Năm 2021, doanh thu thuần về bán hàng và cung cấp dịch vụ tăng mạnh, đạt 9,812 tỷ (tăng 46.93% YOY) do ghi nhận từ các dự án EPC điện gió và mảng thủy điện. Lợi nhuận gộp đạt 1144 tỷ (giảm 1.46% YOY) do thiếu đóng góp từ mảng bất động sản và biên lợi nhuận gộp của mảng xây lắp điện thấp. Lợi nhuận sau thuế (LNST) đạt 764 tỷ (tăng 40% YOY) do ghi nhận doanh thu từ hoạt động tài chính 319 tỷ (tăng 762% YOY) do đánh giá lại khoảng đầu tư tài chính dài hạn vào CTCP khoáng sản Tấn Phát vào Q2/2021.

* Chi phí bán hàng đạt 56 tỷ (giảm 21% YOY) do năm 2020 cần tăng chi phí bán hàng để tiêu thụ dự án PCC1 Thanh Xuân, năm 2021 dự án này chỉ còn 5% còn lại nên chi phí bán hàng giảm đáng kể.

*	Chi phí quản lý doanh nghiệp tăng mạnh từ năm 2020, đặc biệt năm 2020 tăng 33% so với năm 2019 do khoản lợi thế thương mại 281 tỷ phát sinh khi PC1 mua 3 công ty dự án điện gió, trích khấu hao lợi thế thương mại dự kiến trong 4 năm.
*	Mảng sản xuất điện trong năm 2021 đóng góp 9.44% vào doanh thu và 45.83% vào lợi nhuận gộp nhờ hai nhà máy thủy điện đóng điện vào quý 3/2020 (công suất 54MW) đảm bảo doanh thu cho năm 2021. Biên lợi nhuận gộp từ mảng này luôn được duy trì khá cao ở mức 58% đến 62%. Bên cạnh đó, đây cũng là mảng có đóng góp chính vào Lợi nhuận gộp từ năm 2018 nhờ tăng trưởng doanh thu và lợi nhuận gộp bền vững.


Picture5.png

*	Mảng xây lắp điện trong năm 2021, doanh thu đạt 6713 tỷ đồng đóng góp 68.31% vào doanh thu, và đóng góp 39.90% vào lợi nhuận gộp nhờ vào hợp đồng ký mới và lượng backlog 3231 tỷ đồng chuyển từ năm 2020 sang. 
*	Mảng bất động sản, năm 2021 PC1 ghi nhận 5% doanh thu còn lại của dự án PCC1 Thanh Xuân (khoảng 117 tỷ đồng), doanh thu đạt 79 tỷ, đóng góp 1.08% vào doanh thu và 3.34% vào lợi nhuận gộp. Việc thiếu hụt từ doanh thu mảng bất động sản khiến LNST tăng trưởng 40% thấp hơn mức 45% năm 2020. Hiện PC1 đang triển khai các dự án là Vĩnh Hưng, Bắc Từ Liêm, Định Công và Gia Lâm.
*	Các mảng còn lại như sản xuất công nghiệp, bán hàng hóa, vật tư lần lượt đóng góp 6.61% và 14.56% vào doanh thu và 8.10% và 2.83% vào lợi nhuận gộp. Mảng sản xuất công nghiệp, công ty sẽ nâng công suất sản xuất thêm 50% lên 20,000 tấn/năm, với tính hình giá nguyên vật liệu tiếp tục tăng, mảng này có thể giảm đóng góp vào lợi nhuận gộp năm 2022.


##TRIỂN VỌNG NĂM 2022 CHO PC1:

*	Giá khí tăng có thể hỗ trợ công suất các dự án điện gió của PC1. Giá khí tăng đẩy giá bán điện khí lên ít nhất 2000 đồng/kWh trong khi giá điện gió trên bờ là 1928 đồng/kWH và giá bán điện thương phẩm bình quân không thay đổi áp dụng từ 20/03/2019 là 1864 đồng/kWH. Do đó, trước thực trạng giá điện mua vào có thể vượt giá bán ra, EVN có thể phải tìm kiếm nguồn năng lượng rẻ hơn để thay thế khi giá khí tăng. Qua đó nhóm điện gió sẽ được hưởng lợi.

PC1 đầu tư 3 dự án điện gió với công suất và tỷ lệ sở hữu như bảng sau:



image.png

Cả 3 nhà máy Liên Lập, Phong Huy, Phong Nguyên đều đã kịp thời đóng điện và hòa vào lưới điện quốc gia trước thời hạn 31/10/2021 để hưởng cơ chế giá điện hỗ trợ ở mức 8.5 cent/kWh. Điều này kỳ vọng sẽ làm tăng doanh thu mảng điện năm 2022.
*	Bất động sản bắt đầu đóng góp từ năm 2022 nhờ các dự án (1) dự án BDS Định Công và Gia Lâm ước tính ghi nhận doanh thu vào nửa cuối năm 2022, (2) dự án chung cư Vĩnh Hưng sẽ bàn giao và ghi nhận doanh thu từ giữa năm 2022, (3) dự án Bắc Từ Liêm đã hoàn thành đền bù giải phóng mặt bằng, dự kiến bàn giao 2023.
Thêm vào đó, T11/2021 HĐQT đã thông qua việc đóng góp vốn 119 tỷ vào CTCP Thiết bị Thương mại (PC1 sở hữu 98.74%), kỳ vọng công ty sẽ có thêm các dự án từ quỹ đất 1.5 ha tại quận Thanh Xuân, Hà Nội từ công ty con này.
*	Kỳ vọng mảng kim loại tạo đột biến doanh thu và lợi nhuận năm 2022 nhờ giá Niken tăng mạnh do căng thẳng địa chính trị Nga-Ukraine khiến giá các mặt hàng tăng. Giá Niken ngày 14/4/2022 là 33,193 USD/tấn (tăng 24.60% so với tháng 03 – 25,028 USD/tấn). Từ đó, dự án Niken của PC1 dự kiến vận hành Q4/2022 kỳ vọng được hưởng lợi và đóng góp doanh thu, lợi nhuận đột biến khi giá Niken được duy trì ở mức cao. 


###Về trung hạn:
KHẢ QUAN với các luận điểm đầu tư sau:
*	PC1 là nhà thầu uy tín trong mảng xây dựng nhà máy năng lượng tái tạo, theo Quy hoạch điện VIII, công suất điện năng lượng tái tạo sẽ tiếp tục mở rộng trong những năm tiếp theo, do đó khối lượng xây lắp điện cho PC1 sẽ ngày càng tiềm năng, mảng xây lắp sẽ được hưởng lợi từ nhu cầu EPC dự án xây dựng nguồn điện và lưới điện trong trung - dài hạn.
*	Ba dự án điện gió của PC1 hoàn thành đóng điện trước 31/10/2021 để hưởng giá FIT (8,5cent/kWH) trong 20 năm vòng đời theo quyết định của Chính phủ.
